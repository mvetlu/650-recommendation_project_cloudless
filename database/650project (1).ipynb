```python
df = pd.read_csv('/content/initial_amazon_reviews.csv')
df.info()
```

```python
df[df.duplicated()]
```

```python
df.isnull().sum()
df.dropna(inplace=True)
```

```python
# rename AKM1MP6P0OYPR to "User ID", 0132793040 to "Product. ID", 5.0 to "Rating", 1365811200 to "Tiemstamp"
df.rename(columns={'AKM1MP6P0OYPR': 'User_ID', '0132793040': 'Product_ID', '5.0': 'Rating', '1365811200': 'Timestamp'}, inplace=True)
df
```

```python
df['Timestamp'] = df['Timestamp'].astype(float).astype(int)
```

```python
df.head()
```

```python
#save as csv
df.to_csv('amazon_reviews.csv', index=False)
```

```python
#create a new dataset with 10k most active users
sample = df[df['User_ID'].isin(df['User_ID'].value_counts().head(10).index)]
sample.info()
```

```python
sample.reset_index(drop=True, inplace=True)
sample.head()
```

```python
sample.to_csv('interactions_sample.csv', index=False)
```

```python
sample['Product_ID'].nunique()
```
